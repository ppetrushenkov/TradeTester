```python
import pandas as pd

df = pd.read_csv('eurusd_h1.csv', parse_dates=['dt'], index_col=0)
df.drop('pair', axis=1, inplace=True)
df.head()
```

```python
df['hour'] = df.index.hour
```

```python
h1 = df.resample('24h', offset=8, closed='left')
```

```python
h1.last()
```

```python
h1_data = h1.last()
```

```python
h = h1_data[h1_data['hour']==8]
```

```python
h
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=[go.Candlestick(x=h.index,
                open=h.open,
                high=h.high,
                low=h.low,
                close=h.close)])
fig.update_xaxes(
    rangebreaks=[
        { 'pattern': 'day of week', 'bounds': [6, 1]},
        { 'pattern': 'hour', 'bounds':[23,11]}
    ]
)
fig.update_yaxes(title="Price", autorange=True)
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```
